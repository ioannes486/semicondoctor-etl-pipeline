# Extraction - S3 bucket에서 파일 추출

```python
# minIO저장소 접근을 위한 변수 설정
access_key = "access_key"
secret_key = "secret_key"

# 스파크 및 하둡 설정
spark.sparkContext._jsc.hadoopConfiguration().set('fs.s3a.access.key', access_key)
spark.sparkContext._jsc.hadoopConfiguration().set('fs.s3a.secret.key', secret_key)

# file fetch후 Dataframe으로 변환
bucket_name = "my_bucket_name"
file_path = "file_path"
df = spark.read.parquet(f"s3a://{bucket_name}/{file_path}")
```

```python
# 변환된 모습 확인
df.show()
```
